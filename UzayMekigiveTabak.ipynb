```python
import os, shutil

original_dataset_dir1 = 'C:/Users/Merve/Desktop/deeplearning/tabak' #tabak için orijinal veri setinin yolu
original_dataset_dir2 = 'C:/Users/Merve/Desktop/deeplearning/uzaymekigi' #uzay mekiği için orijinal veri setinin yolu

#veri setinin tarin, validation ve test olarak ayrılacağı dosyaların yolunu belirttim.
base_dir = 'C:/Users/Merve/Desktop/deeplearning/tabak_ve_uzay_mekigi' 
os.mkdir(base_dir) #veri setlerini oluşturdum.

train_dir = os.path.join(base_dir, 'train') #train_dir değişkenine train dosyalarını oluşturması için yolunu atadım. 
os.mkdir(train_dir) #train_dir yolunda train adında dosya oluşturdum.
validation_dir = os.path.join(base_dir, 'validation') #validation_dir değişkenine validation dosyalarını oluşturmak için dosyaların yolunu atadım.
os.mkdir(validation_dir) #validation_dir yolunda validation adında dosya oluşturdum.
test_dir = os.path.join(base_dir, 'test') #test_dir değişkenine test dosyalarını oluşturması için dosyaların yolunu atadım.
os.mkdir(test_dir) #test_dir yolunda test adında dosya oluşturdum.

#train_dir yolunda tabak için dosya oluşturdum.
train_tabak_dir = os.path.join(train_dir, 'tabak')
os.mkdir(train_tabak_dir)

#train_dir yolunda uzay mekiği için dosya oluşturdum.
train_uzaymekigi_dir = os.path.join(train_dir, 'uzaymekigi')
os.mkdir(train_uzaymekigi_dir)

#validation_dir yolunda tabak için dosya oluşturdum.
validation_tabak_dir = os.path.join(validation_dir, 'tabak')
os.mkdir(validation_tabak_dir)

#validation_dir yolunda uzay mekiği için dosya oluşturdum.
validation_uzaymekigi_dir = os.path.join(validation_dir, 'uzaymekigi')
os.mkdir(validation_uzaymekigi_dir)

#test_dir yolunda tabak için dosya oluşturdum.
test_tabak_dir = os.path.join(test_dir, 'tabak')
os.mkdir(test_tabak_dir)

#test_dir yolunda uzay mekiği için dosya oluşturdum.
test_uzaymekigi_dir = os.path.join(test_dir, 'uzaymekigi')
os.mkdir(test_uzaymekigi_dir)

#tabak klasörünün içindeki verilerin 0'dan 99'a kadar olan verileri train_dir yolundaki tabak klasörüne kopyaladım.
fnames = ['{}.jpg'.format(i) for i in range(100)]
for fname in fnames:
    src = os.path.join(original_dataset_dir1, fname)
    dst = os.path.join(train_tabak_dir, fname)
    shutil.copyfile(src, dst)
    
#tabak klasörünün içindeki verilerin 100'den 119'a kadar olan verileri validation_dir yolundaki tabak klasörüne kopyaladım.    
fnames = ['{}.jpg'.format(i) for i in range(100, 120)]
for fname in fnames:
    src = os.path.join(original_dataset_dir1,  fname)
    dst = os.path.join(validation_tabak_dir, fname)
    shutil.copyfile(src, dst)
    
#tabak klasörünün içindeki verilerin 120'den 149'a kadar olan verileri test_dir yolundaki tabak klasörüne kopyaladım.    
fnames = ['{}.jpg'.format(i) for i in range(120, 150)]
for fname in fnames:
    src = os.path.join(original_dataset_dir1, fname)
    dst = os.path.join(test_tabak_dir, fname)
    shutil.copyfile(src, dst)
    
#uzaymekigi klasörünün içindeki verilerin 0'dan 99'a kadar olan verileri train_dir yolundaki uzaymekigi klasörüne kopyaladım.    
fnames = ['{}.jpg'.format(i) for i in range(100)]
for fname in fnames:
    src = os.path.join(original_dataset_dir2, fname)
    dst = os.path.join(train_uzaymekigi_dir, fname)
    shutil.copyfile(src, dst)
    
#uzaymekigi klasörünün içindeki verilerin 100'den 119'a kadar olan verileri validation_dir yolundaki uzaymekigi klasörüne kopyaladım.    
fnames = ['{}.jpg'.format(i) for i in range(100, 120)]
for fname in fnames:
    src = os.path.join(original_dataset_dir2, fname)
    dst = os.path.join(validation_uzaymekigi_dir, fname)
    shutil.copyfile(src, dst)
    
#uzaymekigi klasörünün içindeki verilerin 120'den 149'a kadar olan verileri test_dir yolundaki uzaymekigi klasörüne kopyaladım.    
fnames = ['{}.jpg'.format(i) for i in range(120, 150)]
for fname in fnames:
    src = os.path.join(original_dataset_dir2, fname)
    dst = os.path.join(test_uzaymekigi_dir, fname)
    shutil.copyfile(src, dst)
```

```python
#ağ oluşturma
from keras import layers
from keras import models

model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation = 'relu',
                       input_shape = (150, 150, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation = 'relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation = 'relu'))
model.add(layers.MaxPooling2D(2, 2))
model.add(layers.Conv2D(128, (3, 3), activation = 'relu'))
model.add(layers.MaxPooling2D(2, 2))
model.add(layers.Flatten())
model.add(layers.Dense(512, activation = 'relu'))
model.add(layers.Dense(1, activation = 'sigmoid'))
```

```python
model.summary()
```

```python
#optimizer, loss fonksiyonu ve başarım metriği seçtik
from keras import optimizers

model.compile(loss = 'binary_crossentropy',
             optimizer=optimizers.RMSprop(lr=1e-4),
             metrics = ['acc']) 

#.jpg dosyalarımızı 150*150 piksel olacak şekilde ayarladık
from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale = 1./255)
test_datagen = ImageDataGenerator(rescale = 1./255)

train_generator = train_datagen.flow_from_directory(
                    train_dir,
                    target_size = (150, 150),
                    batch_size = 20,
                    class_mode = 'binary')

validation_generator = test_datagen.flow_from_directory(
                        validation_dir,
                        target_size = (150, 150),
                        batch_size = 20,
                        class_mode = 'binary')
```

```python
#batch'leri, tensörleri ve vektörleri görüntüleme
for data_batch, labels_batch in train_generator:
    print('data batch shape:', data_batch.shape)
    print('labels batch shape:', labels_batch.shape)
    break
```

```python
#modelin eğitilmesi
history = model.fit_generator(
                train_generator,
                steps_per_epoch = 20,
                epochs = 20,
                validation_data = validation_generator,
                validation_steps = 20)
```

```python
import matplotlib.pyplot as plt

acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(1, len(acc) + 1)

plt.plot(epochs, acc, 'bo', label = 'Training acc')
plt.plot(epochs, val_acc, 'b', label = 'Validation acc')
plt.title('Training and Validation Accuracy')
plt.legend()

plt.figure()

plt.plot(epochs, loss, 'bo', label = 'Training loss')
plt.plot(epochs, val_loss, 'b', label = 'Validation loss')
plt.title('Training and Validation Loss')
plt.legend()

plt.show()
```

Yaklaşık olarak 300 veri ile train ettiğimiz modelimiz underfit olmuş diyebiliriz, validation ve train değerlerinin birbirinden çok farklı olmasının nedeni verisetinin büyüklüğünün az olmasıdır. Overfit yok.

```python
#Data augmentation
datagen = ImageDataGenerator(
                            rotation_range=40, #resimlerin yönlerini rastgele değiştirdik.
                            width_shift_range=0.2, #yatay olarak rastgele çevirme 
                            height_shift_range=0.2, #dikey olarak rastgele çevirme
                            shear_range=0.2, #rastgele kesme dönüşümleri
                            zoom_range=0.2, #resimleri yakınlaştırma
                            horizontal_flip=True, #yatay asimetri
                            fill_mode='nearest') #bir döndürme veya yakınlaştırmadan sonra yeni oluşan pikselleri doldurmak için 
```

```python
import matplotlib.pyplot as plt
from keras.preprocessing import image

fnames = [os.path.join(train_uzaymekigi_dir, fname) for fname in os.listdir(train_uzaymekigi_dir)]

#data augment için resim seçtik
img_path = fnames[10]

#resmi okuduk ve resize ettik(150*150 olacak şekilde)
img = image.load_img(img_path, target_size = (150, 150))

#numpy dizisine dönüştürdük
x = image.img_to_array(img)

x = x.reshape((1,) + x.shape)

#sonsuza kadar bir döngü olmaması için bir yerde break ekliyoruz.
i = 0
for batch in datagen.flow(x, batch_size = 1):
    plt.figure()
    imgplot = plt.imshow(image.array_to_img(batch[0]))
    i += 1
    if i % 4 == 0:
        break
plt.show()
```

```python
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu',
                        input_shape=(150, 150, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())
model.add(layers.Dropout(0.5))
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer=optimizers.RMSprop(lr=1e-4),
              metrics=['acc'])
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,)

# validation için ayrılan verilerde data augmentation kullanmadık.
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
        train_dir,
        target_size=(150, 150),
        batch_size=32,
        class_mode='binary')

validation_generator = test_datagen.flow_from_directory(
        validation_dir,
        target_size=(150, 150),
        batch_size=32,
        class_mode='binary')

history = model.fit_generator(
      train_generator,
      steps_per_epoch=20,
      epochs=20,
      validation_data=validation_generator,
      validation_steps=20)
```

```python
acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.legend()

plt.figure()

plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()

plt.show()
```
